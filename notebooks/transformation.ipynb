```python
"""
Rascunho da transformação de dados

"""

import json
import pandas as pd

# Carrega os dados brutos
with open("../data/raw/pokemon_raw_json", "r", encoding="utf-8") as f:
    data = json.load(f)

#Verifica a quantidade de registros
len(data)

#Cria a tabela principal
pokemon_df = pd.DataFrame([
    {
        "pokemon_id": p["id"],
        "name": p["name"],
        "height": p["height"],
        "weight": p["weight"],
        "base_experience": p["base_experience"]
    }

    for p in data
])

#Visualização inicial dos dados
pokemon_df.head()

#Informações da tabela
#pokemon_df.info()

#Estatísticas básicas 
pokemon_df.describe()

#Tipos 
data[0]["types"]

pokemon_types = []

for p in data:
    pokemon_id = p["id"]

    for t in p["types"]:
        pokemon_types.append({
            "pokemon_id": pokemon_id,
            "type_name": t["type"]["name"]
        })

pokemon_types_df = pd.DataFrame(pokemon_types)

pokemon_types_df.head()

types_df = pd.DataFrame({
    
    "type_name": pokemon_types_df["type_name"].unique()
})

types_df["type_id"] = types_df.index + 1

types_df.head(17)

pokemon_types_df = pokemon_types_df.merge(
    types_df,
    on="type_name",
    how="left"
)

pokemon_types_df = pokemon_types_df[
    ["pokemon_id", "type_id"]
]

pokemon_types_df.head()

#Abilidades naturais dos pokemon
data[0]["abilities"]

pokemon_abilities = []

for p in data:
    pokemon_id = p["id"]

    for a in p["abilities"]:
        pokemon_abilities.append({
            "pokemon_id": pokemon_id,
            "ability_name": a["ability"]["name"],
            "is_hidden": a["is_hidden"]
        })

pokemon_abilities_df = pd.DataFrame(pokemon_abilities)

pokemon_abilities_df.head()

abilities_df = pd.DataFrame({
    "ability_name": pokemon_abilities_df["ability_name"].unique()
})

abilities_df["ability_id"] = abilities_df.index + 1


abilities_df.head()

pokemon_abilities_df=pokemon_abilities_df.merge(
    abilities_df,
    on = "ability_name",
    how = "left"

)

pokemon_abilities_df = pokemon_abilities_df[
    ["pokemon_id", "ability_id", "is_hidden"]
]

pokemon_abilities_df.head()

#Stats
data[0]["stats"]

pokemon_stats = []

for p in data:
    pokemon_id = p["id"]

    for s in p["stats"]:
        pokemon_stats.append({
            "pokemon_id": pokemon_id,
            "stat_name": s["stat"]["name"],
            "base_stat": s["base_stat"]
        })

pokemon_stats_df = pd.DataFrame(pokemon_stats)

pokemon_stats_df.head()


stats_df = pd.DataFrame({
    "stat_name": pokemon_stats_df["stat_name"].unique()
})

stats_df["stats_id"] = stats_df.index + 1 

stats_df.head()

pokemon_stats_df = pokemon_stats_df.merge(
    stats_df,
    on = "stat_name",
    how="left"
)

pokemon_stats_df = pokemon_stats_df[
    ["pokemon_id", "stats_id", "base_stat"]
]

pokemon_stats_df.head()


#Exportando tabelas processadas para csv

pokemon_df.to_csv("../data/processed/pokemon.csv", index=False)

types_df.to_csv("../data/processed/types.csv", index=False)

pokemon_types_df.to_csv("../data/processed/pokemon_types.csv", index=False)

abilities_df.to_csv("../data/processed/abilities.csv", index=False)

pokemon_abilities_df.to_csv("../data/processed/pokemon_abilities.csv", index=False)

stats_df.to_csv("../data/processed/stats.csv", index=False)

pokemon_stats_df.to_csv("../data/processed/pokemon_stats.csv", index=False)

print("Dados exportados com sucesso!")
```
